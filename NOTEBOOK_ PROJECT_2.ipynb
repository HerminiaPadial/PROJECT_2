```python
#IMPORT

import pandas as pd
import numpy as np
import csv
from unidecode import unidecode
import json
from ftfy import fix_text
```

```python
#EXTRACCIÓN DE DATOS
df = pd.read_csv('./Data/centros2.csv', delimiter=';', encoding='utf-8')
```

```python
df.info()
```

```python
#TABLAS DE DIMENSIONES
```

```python
#DIMENSIÓN1: tipo_centro
df_tipo_centro = df[['centro_tipo_codigo','centro_tipo_desc_abreviada','centro_tipo_descripcion' ]]
df_tipo_centro= df_tipo_centro.drop_duplicates().reset_index(drop=True)
df_tipo_centro=df_tipo_centro.rename(columns={'centro_tipo_codigo':'id_centro_codigo'})
df_tipo_centro
```

```python
#DIMENSIÓN2: TITULARIDAD
df_titularidad = df[['centro_titularidad']]
df_titularidad= df_titularidad.drop_duplicates()
df_titularidad= df_titularidad.reset_index(drop=True)
df_titularidad.reset_index(level=0, inplace=True)
df_titularidad['index']+=1
df_titularidad=df_titularidad.rename(columns={'index':'id_centro_titularidad'})
df_titularidad
```

```python
#MERGE DEL CENTRO_TITULARIDAD CON EL ORIGINAL. 
#pendiente elimiar columna centro_titular del original
df= pd.merge(df, df_titularidad, on="centro_titularidad")   #Se crea columna id_centro_titularidad sobre el dforig
df.info()

```

```python
#DIMENSIÓN3: CENTRO_TITULAR
df_centro_titular= df[['centro_titular']]
df_centro_titular= df_centro_titular.drop_duplicates()
df_centro_titular= df_centro_titular.reset_index(drop=True)
df_centro_titular.reset_index(level=0, inplace=True)
df_centro_titular['index']+=1
df_centro_titular=df_centro_titular.rename(columns={'index':'id_centro_titular'})
df_centro_titular
```

```python
df= pd.merge(df, df_centro_titular, on="centro_titular")
df.info()
```

```python
#DIMENSIÓN4: DAT
df_dat= df[['dat_codigo','dat_nombre']]
df_dat= df_dat.drop_duplicates()
df_dat= df_dat.reset_index(drop=True)
df_dat=df_dat.rename(columns={'dat_codigo':'id_dat_codigo'})
df_dat
```

```python
#DIMENSIÓN5: DF_MUNICIPIOS
df_municipios= df[['municipio_codigo','municipio_nombre']]
df_municipios= df_municipios.drop_duplicates()
df_municipios= df_municipios.reset_index(drop=True)
df_municipios.reset_index(level=0, inplace=True)
df_municipios['index']+=1
df_municipios=df_municipios.rename(columns={'index':'id_municipio'})
df_municipios
```

```python
df= pd.merge(df, df_municipios, on="municipio_codigo")   #Se crea columna id_municipios sobre el dforig
df.info()
```

```python
#DIMENSIÓN6: DF_CP
df_cp= df[['direccion_codigo_postal']]
df_cp= df_cp.drop_duplicates()
df_cp= df_cp.reset_index(drop=True)
df_cp.reset_index(level=0, inplace=True)
df_cp['index']+=1
df_cp=df_cp.rename(columns={'index':'id_direccion_codigo_postal'})
df_cp
```

```python
print(df.columns)
```

```python
df = pd.merge(df, df_tipo_centro, left_on='centro_tipo_codigo', right_on='id_centro_codigo', how='left')
```

```python
df = pd.merge(df, df_dat, left_on='dat_codigo', right_on='id_dat_codigo', how='left')
```

```python
df = pd.merge(df, df_cp, left_on='direccion_codigo_postal', right_on='id_direccion_codigo_postal', how='left')
```
