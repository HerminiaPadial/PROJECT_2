```python
#IMPORT

import pandas as pd
import numpy as np
import csv
from unidecode import unidecode
import json
from ftfy import fix_text
```

```python
#EXTRACCIÓN DE DATOS
df = pd.read_csv('./Data/centros2.csv', delimiter=';', encoding='utf-8')
```

```python
df.info()
```

```python
def combinar_columnas(row):
    valores = [str(row['contacto_telefono1']), str(row['contacto_telefono2']),\
           str(row['contacto_telefono3']), str(row['contacto_telefono4'])]
    valores_validos = [valor for valor in valores if valor != 'nan']
    return '-'.join(valores_validos)
```

```python
df['telefonos'] = df.apply(lambda row: combinar_columnas(row), axis=1)
df = df.drop(['contacto_telefono1', 'contacto_telefono2', 'contacto_telefono3','contacto_telefono4'], axis=1)
```

```python
df.info()
```

```python
df['contacto_web'].fillna('No hay página web', inplace=True)
df['contacto_email1'].fillna('No hay página web', inplace=True)
df['Comunidad'] = 'Comunidad de Madrid'
df['latitud_CM'] = [40.4165000]* len(df)
df['longitud_CM'] = [-3.7025600]* len(df)
```

```python
df.head(3)
```

```python
df.info()
```

```python
#DATA SET 2
#df2 = pd.read_csv('./Data/renta.csv', delimiter=';', encoding='utf-8')
df2 = pd.read_csv('./Data/renta.csv', delimiter=';', encoding='latin-1')
df2.info()
```

```python
df2[['CodigoPostal', 'Nombre']] = df2['Municipios'].str.split(' ', 1, expand=True)
df2['CodigoPostal'] = df2['CodigoPostal'].str.strip()
df2['Nombre'] = df2['Nombre'].str.strip()
df2['Nombre'] = df2['Nombre'].str.replace(r'\s*,{2,}\s*', ' ').str.strip()
df2 = df2.drop(columns=['Municipios', 'Distritos', 'Secciones', 'Periodo', 'Total'])
df2=df2.rename(columns={'Nombre':'municipio_nombre', 'Indicadores de renta media y mediana':'Renta'})
df2.head(4)
```

```python
# Realizar la unión de los DataFrames por la columna 'municipio_nombre'
df_merged = pd.merge(df, df2[['municipio_nombre', 'Renta']], on='municipio_nombre', how='left')

# Verificar y llenar los valores nulos en la columna 'renta' del DataFrame resultante
df_merged['Renta'].fillna(0, inplace=True)  # Reemplazar los valores nulos con 0 o cualquier valor deseado
```

```python
#DIMENSIÓN1: tipo_centro
df_tipo_centro = df_merged[['centro_tipo_codigo','centro_tipo_desc_abreviada','centro_tipo_descripcion' ]]
df_tipo_centro= df_tipo_centro.drop_duplicates().reset_index(drop=True)
df_tipo_centro=df_tipo_centro.rename(columns={'centro_tipo_codigo':'id_centro_codigo'})
df_tipo_centro
```

```python
#DIMENSIÓN2: TITULARIDAD
df_titularidad = df[['centro_titularidad']]
df_titularidad= df_titularidad.drop_duplicates()
df_titularidad= df_titularidad.reset_index(drop=True)
df_titularidad.reset_index(level=0, inplace=True)
df_titularidad['index']+=1
df_titularidad=df_titularidad.rename(columns={'index':'id_centro_titularidad'})
df_titularidad
```

```python
#MERGE DEL CENTRO_TITULARIDAD CON EL ORIGINAL. 
df_merged= pd.merge(df_merged, df_titularidad, on="centro_titularidad")   #Se crea columna id_centro_titularidad sobre el dforig
df_merged.info()

```

```python
#DIMENSIÓN3: CENTRO_TITULAR
df_centro_titular= df_merged[['centro_titular']]
df_centro_titular= df_centro_titular.drop_duplicates()
df_centro_titular= df_centro_titular.reset_index(drop=True)
df_centro_titular.reset_index(level=0, inplace=True)
df_centro_titular['index']+=1
df_centro_titular=df_centro_titular.rename(columns={'index':'id_centro_titular'})
df_centro_titular
```

```python
df_merged= pd.merge(df_merged, df_centro_titular, on="centro_titular")
df_merged.info()
```

```python
#DIMENSIÓN4: DAT
df_dat= df_merged[['dat_codigo','dat_nombre']]
df_dat= df_dat.drop_duplicates()
df_dat= df_dat.reset_index(drop=True)
df_dat=df_dat.rename(columns={'dat_codigo':'id_dat_codigo'})
df_dat
```

```python
#DIMENSIÓN5: DF_MUNICIPIOS
df_municipios= df_merged[['municipio_codigo','municipio_nombre']]
df_municipios= df_municipios.drop_duplicates()
df_municipios= df_municipios.reset_index(drop=True)
df_municipios.reset_index(level=0, inplace=True)
df_municipios['index']+=1
df_municipios=df_municipios.rename(columns={'index':'id_municipio'})
df_municipios
```

```python
df_merged= pd.merge(df_merged, df_municipios, on="municipio_codigo")   #Se crea columna id_municipios sobre el dforig
df_merged.info()
```

```python
#DIMENSIÓN6: DF_CP
df_cp= df_merged[['direccion_codigo_postal']]
df_cp= df_cp.drop_duplicates()
df_cp= df_cp.reset_index(drop=True)
df_cp.reset_index(level=0, inplace=True)
df_cp['index']+=1
df_cp=df_cp.rename(columns={'index':'id_direccion_codigo_postal'})
df_cp
```

```python
#DIMENSIÓN7: DF_comunidad
df_comunidad=df[['Comunidad', 'latitud_CM','longitud_CM']]
df_comunidad= df_comunidad.drop_duplicates()
#df_comunidad= df_comunidad.reset_index(drop=True)
#df_comunidad.reset_index(level=0, inplace=True)
#df_comunidad['index']+=1
df_comunidad=df_comunidad.rename(columns={'Comunidad':'id_comunidad'})
df_comunidad.info()
```
